### 2. Criando uma coluna de Score contendo uma pontuação de acordo com uma preferência de um cliente:



1.   Meu imóvel alvo precisa possuir em torno de 90 m2, então, quero que imóveis maiores ou menores sejam penalizados. Posso pensar algo entre 80 e 100 m2 como toleráveis;
2.   Gostaria de imóveis de 2 quartos, mais que isso fica muito pouco espaço livre e menos que isso não me atende muito bem;
3.   Prefiro imóveis com elevador
4.   Gosto de imóveis com piscina
5.   Gostaria de pagar até 2000 reais, mas até R$ 2.250 eu posso avaliar



Pensando nisso, podemos mapear um score com as seguintes contribuições (uma das possíveis formas, isto é um problema de otimização):

1.   Size => abs(Size - 90) <= 10
2.   Rooms => Rooms == 2
3.   Elevator => Elevator > 0
4.   Swimming Pool => Swimming Pool == 0
5.   Price => abs(2000 - Price) <= 250






```python
imoveisSP['Score'] = (abs(imoveisSP['Size'] - 90) <= 10).astype(int) + \
                     (imoveisSP['Rooms'] % 2 == 0).astype(int) + \
                     (imoveisSP['Elevator']) + \
                     (imoveisSP['Swimming Pool'] == 1).astype(int) + \
                     (abs(imoveisSP['Price'] - 2000) <= 250).astype(int)
```

### 3. Analisando melhores Scores


```python
# Analisando os 5 melhores scores
imoveisSP.nlargest(5, 'Score')
```

```python
# Selecionando as colunas desejadas, filtrando imóveis com Score igual a 5,
# e organizando os resultados por Score e Preço em ordem decrescente
display(imoveisSP[imoveisSP['Score'].eq(5)] \
        [['Price', 'Size', 'Rooms', 'Elevator', 'Swimming Pool', 'District', 'Score']].sort_values(by=['Score', 'Price'], \
                                                                                      ascending=False))

#Exibindo os dois imóveis com os maiores valores de Score e Price
imoveisSP.nlargest(2, ['Score', 'Price'])
```
